Mengimpor library deque

```
from collections import deque
```

Penambahan sistem Queve dan Stack

```
# Stack untuk back dan forward
back_stack = []
forward_stack = []

# Queue untuk history
history = deque()

current_page = None
```

Membuat halaman yang dikunjungi user

```
def visit(page):
    global current_page
    if current_page:
        back_stack.append(current_page)

    current_page = page
    history.append(page)
    forward_stack.clear()

    print("Mengunjungi:", page)
```

Membuat sistem back untuk kembali halaman yang sebelumnya yang pernah dikunjungi

```
def back():
    global current_page
    if not back_stack:
        print("Tidak ada halaman sebelumnya")
        return

    forward_stack.append(current_page)
    current_page = back_stack.pop()

    print("Kembali ke:", current_page)
```

Membuat sistem foward sebagai perintah untuk kembali ke halaman yang dikunjungi setelah melakukan back

```
def forward():
    global current_page
    if not forward_stack:
        print("Tidak ada halaman selanjutnya")
        return

    back_stack.append(current_page)
    current_page = forward_stack.pop()

    print("Maju ke:", current_page)
```

Membuat sistem history

```
def show_history():
    print("\nDaftar History:")
    for page in history:
        print("-", page)
```

Menambahkan simulasi sistem perintah yang ingin dilakukan saat mengunjungi sebuah situs

```
while True:
    print("\n=== Browser History ===")
    print("1. Visit Website")
    print("2. Back")
    print("3. Forward")
    print("4. Lihat History")
    print("5. Keluar")

    choice = input("Pilih menu: ")
```

Penambahan sistem agar perintah yang kita buat sebelumnya dapat bekerja

```
    if choice == "1":
        page = input("Masukkan nama website: ")
        visit(page)

    elif choice == "2":
        back()

    elif choice == "3":
        forward()

    elif choice == "4":
        show_history()

    elif choice == "5":
        print("Program selesai")
        break

    else:
        print("Pilihan tidak valid")
```

Ketika semua code disatukan maka akan membuat suatu halaman

```
from collections import deque

# Stack untuk back dan forward
back_stack = []
forward_stack = []

# Queue untuk history
history = deque()

current_page = None

def visit(page):
    global current_page
    if current_page:
        back_stack.append(current_page)

    current_page = page
    history.append(page)
    forward_stack.clear()

    print("Mengunjungi:", page)

def back():
    global current_page
    if not back_stack:
        print("Tidak ada halaman sebelumnya")
        return

    forward_stack.append(current_page)
    current_page = back_stack.pop()

    print("Kembali ke:", current_page)

def forward():
    global current_page
    if not forward_stack:
        print("Tidak ada halaman selanjutnya")
        return

    back_stack.append(current_page)
    current_page = forward_stack.pop()

    print("Maju ke:", current_page)

def show_history():
    print("\nDaftar History:")
    for page in history:
        print("-", page)

while True:
    print("\n=== Browser History ===")
    print("1. Visit Website")
    print("2. Back")
    print("3. Forward")
    print("4. Lihat History")
    print("5. Keluar")

    choice = input("Pilih menu: ")

    if choice == "1":
        page = input("Masukkan nama website: ")
        visit(page)

    elif choice == "2":
        back()

    elif choice == "3":
        forward()

    elif choice == "4":
        show_history()

    elif choice == "5":
        print("Program selesai")
        break

    else:
        print("Pilihan tidak valid")
```
